```python
import pandas as pd 
order=pd.read_csv(r'C:\Users\Shubham\Project\Ecommerce Profitability Analysis\data\orders.csv')
products=pd.read_csv(r'C:\Users\Shubham\Project\Ecommerce Profitability Analysis\data\products.csv')
marketing=pd.read_csv(r'C:\Users\Shubham\Project\Ecommerce Profitability Analysis\data\marketing_spend.csv')

print(f"Order:{order.shape} having columns {order.columns.to_list()}")
print(f"Products:{products.shape} having columns {products.columns.to_list()}")
print(f"Marketing:{marketing.shape} having columns {marketing.columns.to_list()}")

print(f"Order head:{order.head(3)}")
print(f"Products head:{products.head(3)}")
print(f"Marketing head:{marketing.head(3)}")
```

```python
print(f"Order Nulls:{order.isnull().sum()}")
print(f"products Nulls:{products.isnull().sum()}")
print(f"Marketing Nulls:{marketing.isnull().sum()}")

print(f"Order Duplicates:{order.duplicated().sum()}")
print(f"Prodcuts Duplicates:{products.duplicated().sum()}")
print(f"Marketing Duplicates:{marketing.duplicated().sum()}")

print(f"Order Dtypes:{order.dtypes}")
print(f"Products Dtypes:{products.dtypes}")
print(f"marketing Dtypes:{marketing.dtypes}")
```

```python
# Basic Stats
print("---Order stats---")
print(order[[
    'gross_revenue', 'discount_pct', 'shipping_cost','product_cost', 'platform_fee', 'profit','refund_amount'
]].describe())

print("\n --Marketing Stats--")
print(marketing[['spend', 'revenue_attributed','roas', 'cpc', 'cpa']].describe())
```

```python
# ── Unique values in categorical columns ──
print("Channels      :", order['channel'].unique())
print("Categories    :", order['primary_category'].unique())
print("Returned      :", order['returned'].unique())
print("Regions       :", order['region'].unique())
print("Platforms     :", marketing['platform'].unique())
print("Payment methods:", order['payment_method'].unique())
```

```python
order["order_date"]=pd.to_datetime(order['order_date'])

print("Order date range::")
print(f"From:{order['order_date'].min()}")
print(f"To :{order['order_date'].max()}")

print("\n Marketing months")
print(marketing['month'].unique())
```

```python
# Verifing gross_revenue-total_costs=profit

order['profit_check']=(order['gross_revenue']-order['total_costs']).round(2)

mismatch=order[order['profit_check']!=order['profit'].round(2)]
print(f"Profit calculation mismatches:{len(mismatch)}")

returned_with_revenue=order[(order['returned']=='Yes')&(order['net_revenue']>0)]


print(f"Returned orders with positive net revenue:{len(returned_with_revenue)}")
```

```python
# Look  mismatch rows closely
mismatch = order[order['profit_check'] != order['profit'].round(2)].copy()

print(f"Total mismatches: {len(mismatch)}")
print(f"\nSample mismatches:")
print(mismatch[['order_id', 'gross_revenue', 'total_costs', 
 'profit', 'profit_check', 'returned', 'refund_amount', 'discount_amount']].head(10).to_string())


print("Mismatches by returned status:")
print(mismatch['returned'].value_counts())


print("\nNon-returned mismatches:")
non_returned_mismatch = mismatch[mismatch['returned'] == 'No']
print(non_returned_mismatch[['order_id', 'gross_revenue', 'total_costs', 'profit', 'profit_check']].head(5).to_string())



# Check the difference amount
mismatch['difference'] = (mismatch['profit'] - mismatch['profit_check']).round(2)

print("Difference stats:")
print(mismatch['difference'].describe())
print("\nUnique difference values (first 20):")
print(mismatch['difference'].value_counts().head(20))


# Checking if refund amount explains the difference
mismatch['refund_explains'] = (
    mismatch['profit_check'] - mismatch['refund_amount']
).round(2) == mismatch['profit'].round(2)

print("Does refund_amount explain the difference?")
print(mismatch['refund_explains'].value_counts())
```

```python
# Test hypothesis: profit = gross_revenue - total_costs - discount_amount
order['profit_test'] = (
    order['gross_revenue'] - order['total_costs'] - order['discount_amount']
).round(2)

match = (order['profit_test'] == order['profit'].round(2)).sum()
print(f"Matches with discount subtracted: {match} / {len(order)}")

# Test for returned orders specifically
order['profit_test2'] = (
    order['gross_revenue'] - order['total_costs'] - 
    order['discount_amount'] - order['refund_amount']
).round(2)

match2 = (order['profit_test2'] == order['profit'].round(2)).sum()
print(f"Matches with discount + refund subtracted: {match2} / {len(order)}")
```

```python

order['profit_check'] = (
    order['gross_revenue'] - order['total_costs'] - 
    order['discount_amount'] - order['refund_amount']
).round(2)

# V should be 2000/2000 now
match = (order['profit_check'] == order['profit'].round(2)).sum()
print(f"Profit verification: {match} / {len(order)} match")

# Use existing profit column — it's correct
print("\nSample verification:")
print(order[['order_id', 'gross_revenue', 'discount_amount', 
              'refund_amount', 'total_costs', 
              'profit']].head(5).to_string())
```

```python
# ── Negative profit investigation ──
neg_profit = order[order['profit'] < 0]
print(f"\n=== NEGATIVE PROFIT ORDERS ===")
print(f"Count: {len(neg_profit)} / {len(order)}")
print(f"Total loss from negative profit orders: ${neg_profit['profit'].sum():,.2f}")
print("\nNegative profit by category:")
print(neg_profit['primary_category'].value_counts())
```

```python
# ── Nulls ──
print("=== NULLS ===")
print("Order nulls:")
print(order.isnull().sum()[order.isnull().sum() > 0])
print("\nProducts nulls:")
print(products.isnull().sum()[products.isnull().sum() > 0])
print("\nMarketing nulls:")
print(marketing.isnull().sum()[marketing.isnull().sum() > 0])
# ── Duplicates ──
print("=== DUPLICATES ===")
print(f"Order      : {order.duplicated().sum()}")
print(f"Products   : {products.duplicated().sum()}")
print(f"Marketing  : {marketing.duplicated().sum()}")
# ── Categorical checks ──
print("=== UNIQUE VALUES ===")
print("Channels        :", order['channel'].unique())
print("Categories      :", order['primary_category'].unique())
print("Returned        :", order['returned'].unique())
print("Regions         :", order['region'].unique())
print("Platforms       :", marketing['platform'].unique())
print("Payment methods :", order['payment_method'].unique())
# ── Date check ──
print("=== DATE RANGE ===")
print(f"Orders from : {order['order_date'].min().date()}")
print(f"Orders to   : {order['order_date'].max().date()}")
print(f"\nMarketing covers: {marketing['month'].min()} → {marketing['month'].max()}")

future = order[order['order_date'] > '2025-12-31']
print(f"\nOrders after Dec 2025: {len(future)}")
if len(future) > 0:
    print(future['order_date'].value_counts().head())
# ── Key stats ──
print("=== ORDER KEY STATS ===")
print(order[['gross_revenue', 'profit', 'discount_pct',
             'shipping_cost', 'platform_fee',
             'refund_amount']].describe().round(2))
```

```python
# ── Fix date type ──
order['order_date'] = pd.to_datetime(order['order_date'])

# ── Fix marketing month to datetime for time series later ──
marketing['month_dt'] = pd.to_datetime(marketing['month'])

# ── Add profit margin % column ──
order['profit_margin_%'] = (
    order['profit'] / order['gross_revenue'] * 100
).round(2)

# ── Add binary returned column for easier aggregation ──
order['returned_binary'] = (order['returned'] == 'Yes').astype(int)

# ── Add month column for time series joining with marketing ──
order['month'] = order['order_date'].dt.to_period('M').astype(str)

# ── Filter out 2026 orders — marketing data only covers 2024-2025 ──
order_clean = order[order['order_date'] <= '2025-12-31'].copy()

print(f"Original order count  : {len(order)}")
print(f"Clean order count     : {len(order_clean)}")
print(f"Removed (2026 dates)  : {len(order) - len(order_clean)}")
print(f"\nNew columns added:")
print(order_clean[['profit_margin_%', 'returned_binary', 
                    'month']].head(3))
```

```python
order_clean.to_csv('../data/order_clean.csv', index=False)
products.to_csv('../data/products_clean.csv', index=False)
marketing.to_csv('../data/marketing_clean.csv', index=False)

print(" All clean files exported")
print(f"order_clean    : {len(order_clean)} rows")
print(f"products_clean : {len(products)} rows")
print(f"marketing_clean: {len(marketing)} rows")
```
